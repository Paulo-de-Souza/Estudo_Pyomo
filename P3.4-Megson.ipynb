##Função de Tensão Triângulo Equilátero

A equação a seguir é como pode ser descrita a função de tensão para um triângulo equilatero

$$\phi = -G \frac{d \theta}{dz} \left[\frac{1}{2}(x^2+y^2) - \frac{1}{2a}(x^3-3xy^2) - \frac{2}{27}a^2 \right]$$

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

n = 0.01
a = 10
c = 2/27 * a**2
x = np.arange(-15,15,n)
y = np.arange(-15,15,n)
X,Y = np.meshgrid(x,y)
Z = 0.5*(X**2 + Y**2) - (1/(2*a))*(X**3 - 3*X*Y**2)  - c
fig, ax = plt.subplots()

cp = ax.contourf(X,Y,Z,cmap = 'inferno')
#ax.clabel(cp,inline = True,fontsize=10)
plt.axis('equal')
plt.show()
```
